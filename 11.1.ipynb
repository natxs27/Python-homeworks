```python
import numpy as np
import matplotlib.pyplot as plt

D, Nx, Nt = 0.25, 60, 250
dx = 1.0/Nx
dt = 0.1/Nt
r = D*dt/(dx**2)
assert r <= 0.5

x = np.linspace(0.0, 1.5, Nx+1)
u = np.empty((Nx+1, Nt+1), dtype=float)
u[:,0] = -x**2+0.1
u[0,:] = np.linspace(0.5, 0, Nt+1)
u[Nx,:] = np.linspace(0.0, 0.5, Nt+1)
for a in range(Nt):
    for b in range (1,Nx):
        u[b, a+1] = r*u[b-1, a]+(1-2*r)*u[b, a] + r*u[b+1, a]

plt.title('Jednowymiarowe równanie przewodnictwa ciepła')
plt.imshow(u, cmap='afmhot', interpolation='nearest')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
